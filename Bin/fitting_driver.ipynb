```python
import xlrd
import math
import numpy as np
from scipy.optimize import minimize
from Fitter.fitting import fitting
```

### Hagan Approximation

```python
# calibrate the Hagan SABR model to input data
# The variable results is a dictionary which stores the ivols, ivols_diff(ivols-market ivols) and parameters of SABR 
# for unconstraint calibration and constraint calibration with beta/rho/alpha/vega fixed to specific values
results=fitting('market_data.xlsx')
```

```python
auto=results['auto'] # unconstraint calibration results of Hagan SABR model
print ('ivols predicted by Hagan SABR model:\n') 
print (auto['ivols']) # print implied vols
print ('\nivols differences predicted by Hagan SABR model:\n') 
print (auto['ivols_diff']) # print implied vols difference
print ('\nparameters of Hagan SABR model calibrated:\n')
print (auto['params']) # print parameters of Hagan SABR calibrated
```

```python
# Take beta_check as an example
beta_check=results['beta_check'] # constraint calibration results of Hagan SABR model with beta fixed to specific numbers
print (beta_check[0]) # print calibration results when beta is fixed to 0
```
