** Take data of 5Y tenor and 10Y expiry as an example **

```python
fitter=Fitter('market_data.xlsx')
results_hagan=fitter.calibration() #Hagan lognormal implementation
results_obloj=fitter.calibration(method='Obloj') #Obloj implementation
ivols_hagan=fitter.ivol_SABR(results_hagan['alpha'],results_hagan['beta'],results_hagan['rho'],results_hagan['nu'])
ivols_obloj=fitter.ivol_SABR(results_obloj['alpha'],results_obloj['beta'],results_obloj['rho'],results_obloj['nu'])
params_hagan=results_hagan['params']
params_obloj=results_obloj['params']

factors_hagan=params_hagan.loc[13,:][2:]
vols_hagan=ivols_hagan.loc[13,:]
factors_obloj=params_obloj.loc[13,:][2:]
vols_obloj=ivols_obloj.loc[13,:]
print ('Hagan params:\n', factors_hagan.transpose())
print ('Hagan ivols:\n', vols_hagan)
print ('Obloj params:\n', factors_obloj)
print ('Obloj ivols:\n', vols_obloj)
```

```python
forwards_hagan=simulateSABRMonteCarloEuler(100000,40,10,vols_hagan['F'],factors_hagan['alpha'],factors_hagan['beta'],factors_hagan['rho'],factors_hagan['nu'])
forwards_obloj=simulateSABRMonteCarloEuler(100000,40,10,vols_obloj['F'],factors_obloj['alpha'],factors_obloj['beta'],factors_obloj['rho'],factors_obloj['nu'])
print ('Hagan MC standard error:', np.std(forwards_hagan))
print ('Obloj MC standard error:', np.std(forwards_obloj))
```

```python
#plot MC simulated option prices against BS prices with Hagan ivol
p1=[] #MC simulated option price by Hagan params
p2=[] #BS price by Hagan ivol
p3=[] #BS price by Obloj ivol
p4=[] #MC simulated option price by Obloj params
for K_spread in vols_hagan.index.tolist()[3:]:
    option=[max(0, item-(vols_hagan['F']+K_spread/10000.)) for item in forwards_hagan]
    option2=[max(0, item-(vols_obloj['F']+K_spread/10000.)) for item in forwards_obloj]
    p1.append(sum(option)/len(option))
    p4.append(sum(option2)/len(option2))
    p2.append(black(vols_hagan['F'],(vols_hagan['F']+K_spread/10000.),10,vols_hagan[K_spread],1))
    p3.append(black(vols_obloj['F'],(vols_obloj['F']+K_spread/10000.),10,vols_obloj[K_spread],1))  
plt.plot(vols_hagan.index.tolist()[3:],p1,label='MC by Hagan')
plt.plot(vols_hagan.index.tolist()[3:],p4,label='MC by Obloj')
#plt.plot(vols_hagan.index.tolist()[3:],p2,label='Hagan lognormal')
#plt.plot(vols_obloj.index.tolist()[3:],p3,label='Obloj lognormal')
plt.title('MC simulated option price vs predicted price by Hagan lognormal')
plt.xlabel('K_spreads')
plt.ylabel('option price')
plt.legend()
plt.show()
```

```python
#plot MC simulated ivols against Hagan ivols
ivols_hagan=vols.values[3:] #Hagan ivol
ivols_MC=[] #MC simulated ivol
for counter, K_spread in enumerate(vols.index.tolist()[3:]):
    ivols_MC.append(find_ivol(p1[counter],vols['F'],vols['F']+K_spread/10000.,10))
plt.plot(vols.index.tolist()[3:],[item*100 for item in ivols_MC],label='MC')
plt.plot(vols.index.tolist()[3:],[item*100 for item in ivols_hagan],label='Hagan lognormal')
plt.title('MC simulated ivol vs Hagan lognormal ivol (%)')
plt.xlabel('K_spreads')
plt.ylabel('ivol')
plt.legend()
plt.show()
```
