```python
import math
import numpy as np
import pandas as pd
from Pricing.SABR import SABR_model
from Pricing.Data_processor import data_reader,set_label
```

```python
data = data_reader('market_data.xlsx','Swaptions data')
labels = set_label(data['strike_spreads'],data['num_strikes'],data['expiries'],data['tenors'])
```

```python
ATM_vol=[]
ATM_MKT=[]
sabr=SABR_model(0.5,0,0.25) #random nos
for i in range(len(data['F'])):
    for j in range(data['K'].shape[1]):
        if data['F'][i]==data['K'][i][j]:
            vol = sabr.ivol_Hagan_lognorm(0.001,data['F'][i],data['K'][i][j],data['expiries'][i])
            ATM_vol.append(vol)
            ATM_MKT.append(data['MKT'][i][j])
for i in range(len(ATM_vol)):
    print('vol_Hagan=',ATM_vol[i])
    print('MKT=',ATM_MKT[i])
```
